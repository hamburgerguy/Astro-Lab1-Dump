```python
import ugradio
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

# 4.7.1

```python
x = ugradio.pico.capture_data('1V', nsamples=16000)
np.savetxt('data_1.0', x)
```

## Mean Voltage, Mean Square Voltage, and Root-Mean-Square Voltage

```python
x = np.loadtxt('data_1.0')

m = np.mean(x)
ms = np.mean(x**2)
rms = np.sqrt(np.mean(x**2))
print 'Mean Voltage: ', m, ', Mean Square Voltage: ', ms, ', Root-Mean-Square Voltage: ', rms
```

# 4.7.5

## ACF of Block 1

```python
# acf of block 1

x = np.loadtxt('data_1.1')
y = np.arange(1, 16000)
acf = np.correlate(x, y, mode='same')
```

```python
print(acf)
```

```python
# power spectrum derived from acf

f, fx = ugradio.dft.dft(acf[0:15999])
Pacf = fx*np.conj(fx)
```

```python
# power spectrum of block 1

f, fx = ugradio.dft.dft(x[0:15999])
P = fx*np.conj(fx)
```

## FWHM of ACF

```python
from scipy.interpolate import UnivariateSpline

y = Pacf / Pacf.max() * 100
x = np.arange(1, 16000)

spline = UnivariateSpline(x,y-50,s=0)
Tfwhm = abs(spline.roots()[1]-spline.roots()[0])
print(Tfwhm)
```

## FWHM of Power Spectrum

```python
from scipy.interpolate import UnivariateSpline

y = P / P.max() * 100
x = np.arange(1, 16000)

spline = UnivariateSpline(x,y-50,s=0)
Ffwhm = abs(spline.roots()[1]-spline.roots()[0])
print(Ffwhm)
```

The FWHM of the ACF and that of the power spectrum are fairly similar.

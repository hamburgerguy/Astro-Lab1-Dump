## Confirmed sMBHs

```python
mass = [4, 4, 5.2, 6, 6.3, 6.8, 7.0, 7.1, 7.5, 7.6, 7.8, 9.4, 9.6, 10, 11, 12, 14]
error = [[1, 0, 0.6, 2, 0.3, 0.4, 0.6, 0.3, 0.3, 1.3, 0, 1.0, 1.2, 3, 2, 2, 4], # lower
         [1, 16, 0.6, 2, 0.3, 0.4, 0.6, 0.3, 0.3, 1.3, 12.2, 1.0, 1.2, 3, 2, 2, 4]] # upper

x = np.arange(len(mass))

plt.errorbar(x + 1, mass, yerr=error, fmt='k.')
plt.title('Confirmed Stellar-Mass Black Holes')
plt.xticks([])
plt.ylabel('Mass [$M_{\odot}$]')
plt.ylim(0, 20)
plt.show()
```
